```python
from itertools import combinations

import pandas as pd
from rdkit import Chem

import scaffound
```

```python
data = pd.concat((pd.read_csv('../tests/cox2_816_inhibitors.txt', sep='\t'), pd.read_csv(
    '../tests/additional_scaffolds.txt', sep='\t'))).Molecule_SMILES.rename('molecule').apply(Chem.MolFromSmiles)
```

```python
anatomy = data.apply(lambda x: scaffound.MolecularAnatomy(x).to_pandas())
anatomy = anatomy.drop(columns=anatomy.filter(regex=r'^(?:saturated (?!graph))|(?:generic (?!graph))').columns)
```

```python
saturated_anatomy = anatomy['saturated graph'].apply(lambda x: scaffound.MolecularAnatomy(x).to_pandas())
generic_anatomy = anatomy['generic graph'].apply(lambda x: scaffound.MolecularAnatomy(x).to_pandas())
```

```python
full_anatomy = pd.concat((anatomy.add_prefix('original '), generic_anatomy.add_prefix('generic '), saturated_anatomy.add_prefix('saturated ')), axis=1).map(Chem.MolToSmiles)
```

```python
full_anatomy.columns = full_anatomy.columns.str.replace(' ', '_')
full_anatomy = full_anatomy.drop(columns=full_anatomy.filter(regex='saturated_saturated').columns.tolist() + full_anatomy.filter(regex='generic_generic').columns.tolist())
```
